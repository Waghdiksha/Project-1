```python
def calculate_outliers_percentage(df, columns):
    outliers = {}
    for col in columns:
        z = np.abs(stats.zscore(df[col]))
        outlier_count = (z > 3).sum()
        outliers[col] = (outlier_count / len(df[col])) * 100
    return pd.DataFrame(outliers, index=['Outlier Percentage %']).round(3)
```
